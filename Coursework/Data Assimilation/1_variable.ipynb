```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
sigma_o = 1
y = 2

sigma_b = 2
x_b = 1

alpha = sigma_o**2 / (sigma_o**2 + sigma_b**2)
x_a = alpha * y + (1 - alpha) * x_b

# plot pdf with mean y and variance sigma_o^2
x = np.linspace(-5, 5, 100)
plt.plot(x, stats.norm.pdf(x, y, sigma_o), label='y')
plt.plot(x, stats.norm.pdf(x, x_b, sigma_b), label='x_b')
plt.plot(x, stats.norm.pdf(x, x_a, sigma_o), label='x_a')
```
