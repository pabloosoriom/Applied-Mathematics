```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
theta = np.linspace(-10, 10, 100)

theta_o = 3
sigma2_o = 10

theta_b = 2
sigma2_b = 10

# Posterior mean (Analysis)
alpha_a = sigma2_o / (sigma2_o + sigma2_b)
theta_a = alpha_a * theta_b + (1 - alpha_a) * theta_o

# Posterior variance (Analysis)
sigma2_a = sigma2_o * sigma2_b / (sigma2_o + sigma2_b)

pdf_priori = stats.norm.pdf(theta, theta_b, np.sqrt(sigma2_b))
pdf_observation = stats.norm.pdf(theta, theta_o, np.sqrt(sigma2_o))
pdf_posteriori = stats.norm.pdf(theta, theta_a, np.sqrt(sigma2_a))

plt.plot(theta, pdf_priori, label='Priori')
plt.plot(theta, pdf_observation, label='Observation')
plt.plot(theta, pdf_posteriori, label='Posteriori')
plt.grid(True)

plt.xlabel('$\\theta$', fontdict={'size':18});
plt.legend(['$\\mathcal{P}(x )$','$\\mathcal{L}(x \\vert y)$','$\\mathcal{P}(x \\vert y)$'], 
           prop={'size':18});
```
